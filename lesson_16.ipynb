## Генерація датасету і завдання.

- побудувати просту систему рекомендацій

```python
import pandas as pd
import numpy as np
np.random.seed(12)

data = {'userId':np.random.randint(1,20,500),
        'itemId': np.random.randint(101,111,500),
        'rating': np.random.randint(1,6,500)}

df = pd.DataFrame(data=data)
df = df.sort_values(by='rating', ascending=False).reset_index(drop=True)
df.head(10)
```

```python
item_mean = df.groupby('itemId')['rating'].mean().reset_index()
item_mean = item_mean.sort_values(by='rating', ascending=False)
print('Top-5 items to recommend: ')
item_mean.head()
```

```python
# побудова матриці айтемів
df_unique = df.drop_duplicates(subset=['userId', 'itemId'])
R = df_unique.pivot(index='userId', columns='itemId', values='rating').fillna(0)
R
```

```python
user_means = R.replace(0, np.nan).mean(axis=1)
user_means
```

```python
R_norm = R.sub(user_means, axis=0).fillna(0)
R_norm
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(R_norm)
sim_users = pd.DataFrame(similarity_matrix, index=R.index, columns=R.index)

similarity_matrix, sim_users
```

```python
# функція згенерована ШІ
def predict_user_rating(user_id: int, item_id: int, k: int = 3) -> float:
    """
    Прогноз рейтингу для конкретного користувача та предмета за User–User CF.

    Повертає np.nan, якщо користувача немає або якщо немає достатньо інформації.
    У разі відсутності предмета чи валідних сусідів повертає середній рейтинг користувача.
    """
    # Перевірка наявності користувача та предмета
    if user_id not in sim_users.index or user_id not in R.index or user_id not in user_means.index:
        print(f"User {user_id} не знайдено у матриці.")
        return np.nan
    if item_id not in R.columns:
        fallback = user_means.get(user_id, np.nan)
        print(f"Item {item_id} не знайдено. Використано середнє користувача: {fallback:.2f}")
        return fallback

    sims = sim_users.loc[user_id].drop(user_id)
    rated_by = R.index[R[item_id] > 0]
    if len(rated_by) == 0:
        fallback = user_means.get(user_id, np.nan)
        print(f"Немає оцінок для item {item_id}. Використано середнє користувача: {fallback:.2f}")
        return fallback

    valid_neighbors = rated_by.intersection(sims.index)
    if len(valid_neighbors) == 0:
        fallback = user_means.get(user_id, np.nan)
        print(f"Немає доступних сусідів, що оцінили item {item_id}. Використано середнє користувача: {fallback:.2f}")
        return fallback

    neighbors = sims.loc[valid_neighbors].nlargest(k)
    if neighbors.empty:
        fallback = user_means.get(user_id, np.nan)
        print(f"Після відбору сусідів нічого не залишилось. Використано середнє користувача: {fallback:.2f}")
        return fallback

    ratings_diff = R.loc[neighbors.index, item_id] - user_means.loc[neighbors.index]
    num = (ratings_diff * neighbors).sum()
    den = neighbors.abs().sum()
    if den == 0:
        fallback = user_means.get(user_id, np.nan)
        print(f"Сума ваг сусідів = 0. Використано середнє користувача: {fallback:.2f}")
        return fallback

    prediction = user_means[user_id] + num / den
    print(f"Прогноз для user {user_id} на item {item_id}: {prediction:.2f}")
    return prediction
```

```python
t_len = len(df['userId'].unique().tolist())
for k in range(1, t_len):
    _ = predict_user_rating(user_id=k, item_id=102)
```

```python
def get_recommendations(user_id, n=5, k=3):
    """
    Повертає top-n рекомендованих item_id для заданого user_id.
    Використовує просту User–User CF (функція predict_user_rating).
    """
    # Якщо користувача немає в матриці – повертаємо порожній список
    if user_id not in R.index:
        return []

    # Збираємо список всіх item_id, які користувач ще не оцінив
    unrated_items = R.columns[R.loc[user_id] == 0]

    # Для кожного такого item робимо прогноз
    preds = []
    for item_id in unrated_items:
        pred = predict_user_rating(user_id, item_id, k=k)
        # Якщо отримали NaN (наприклад, немає сусідів), пропускаємо
        if np.isnan(pred):
            continue
        preds.append((item_id, pred))

    # Сортуємо за прогнозованим рейтингом (спадання)
    preds.sort(key=lambda x: x[1], reverse=True)

    # Повертаємо лише перші n item_id
    return [item_id for item_id, _ in preds[:n]]

# Наприклад, для user_id = 3 отримати 5 рекомендацій:
recommendations = get_recommendations(user_id=4, n=5, k=3)
print("Рекомендовані item_id:", recommendations)
```
