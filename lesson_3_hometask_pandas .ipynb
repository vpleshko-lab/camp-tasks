### Question 2
What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This function should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

````python
def answer_two():
    avgGDP = merged_df[["2006","2007","2008","2009","2010","2011","2012","2013","2014","2015"]].mean(axis=1,skipna=True)
    avgGDP = avgGDP.sort_values(ascending=False)
    return avgGDP.head(15)
answer_two()
````

### Question 3
By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

*This function should return a single number.*

````python
"""1. Визначити, яка країна займає 6 місце серед середніх за 10 років по ВВП
    2. Ідентифікувати ВВП (France) цієї країни у 2006 та 2015
    3. Відняти значення"""

def answer_three():
    avgGDP = merged_df[["2006","2007","2008","2009","2010","2011","2012","2013","2014","2015"]].mean(axis=1,skipna=True)
    avgGDP = avgGDP.sort_values(ascending=False)
    country6_id,country6_value = avgGDP.index[5],avgGDP[5]

    gdp_2006 = merged_df.loc[country6_id,"2006"]
    gdp_2015 = merged_df.loc[country6_id,"2015"]
    result = gdp_2015 - gdp_2006
    
    print(f"{result} (for {country6_id})")
answer_three()
````

### Question 5

Create a column that estimates the population using Energy Supply and Energy Supply per capita. 
What is the third most populous country according to this estimate?

*This function should return a single string value.*

````python
"""
1. Новий стовпчик -> ["Estimated Population"] 
2. Відсортувати дф за спаданням та ідентифікувати 3-у країну
3. Вивести країну
"""

def answer_five():
    merged_df["Estimated population"]= merged_df["Energy Supply"]/merged_df["Energy Supply per Capita"]
    merged_sorted = merged_df.sort_values(by="Estimated population",ascending=False)
    result = merged_sorted.iloc[2].name
    return result
answer_five()
````

### Question 6
Create a column that estimates the number of citable documents per person. 
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*


````python
"""
Новий стовпець: Citable Docs. per person = "Citable documents"/Estimated population
Використати корреляцію пірсона. Якщо рез: близьке до -1: сильний обернений звязок, до 0: звязку практично немає, близьке до 1: сильний звязок

"""


def answer_six():
    merged_df["Citable Docs. per capita"]= merged_df["Citable documents"]/merged_df["Estimated population"]
    result = merged_df[["Citable Docs. per person","Energy Supply per Capita"]].corr().iloc[0,1]
    return result
answer_six()
````

### Question 7
Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python

def answer_seven():
    test_df = merged_df

    test_df["Continent"] = test_df.index.map(ContinentDict)
    continent_stats = test_df.groupby("Continent")["Estimated population"].agg(
        size="size",
        sum="sum",
        mean="mean",
        std="std"
    )
    return continent_stats

answer_seven()
````
