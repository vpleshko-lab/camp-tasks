## Task 1: Light vs dark theme

```python
path = '/Users/vladislavpleshko/Downloads/Light vs Dark Theme Case Study.csv'
df = pd.read_csv(path)
df
```

```python
print(df.isnull().sum())
df.describe()
```

```python
df['Purchases_enc'] = df['Purchases'].replace({'Yes': 1, 'No': 0})
df
```

```python
df[['Theme', 'Purchases','Purchases_enc']].head()
```

```python
theme_counts = df.groupby('Theme')['Purchases_enc'].count()

# Для цікавості, порахую різницю динаміки між темами
light_mean = theme_counts['Light Theme']
dark_mean = theme_counts['Dark Theme']
percent_diff = (dark_mean - light_mean) / light_mean * 100
print(f"Відсоток різниці між темами: {percent_diff:.2f}%")

theme_counts
```

```python
# формування H0, H1
light_group = df.loc[df['Theme'] == 'Light Theme','Purchases_enc'].values
dark_group = df.loc[df['Theme'] == 'Dark Theme', 'Purchases_enc'].values

print(len(light_group), len(dark_group))
```

```python
t_stat, p_value = stats.ttest_ind(light_group, dark_group, equal_var=False)
print(f'T-stat = {round(t_stat,5)}, P-value = {round(p_value,5)}')

alpha = 0.05
if p_value < alpha:
    print('Відкидаємо H0: частки покупок у групах "light & dark" статистично значущо різняться')
else:
    print('Не відкидаємо H0: статистично значущої різниці у купівельній поведінці між темами не виявлено.')
```

**Висновок**: за наявними даними різниця в частці покупок між Dark Theme і Light Theme (5.76%) не є статистично значущою (p ≈ 0.39). Отже, наразі нема обґрунтованих підстав стверджувати, що саме колірна тема впливає на конверсію.

```python
''' Додатково, мені ще хотілось провести аналіз активності користувача залежно від теми, цікаво було отримати результати.
    Якщо у вас є просунутіші ідеї як це можна було реалізувати - буду радий фідбеку)'''

list_criteria = ['Click Through Rate', 'Conversion Rate', 'Bounce Rate',
       'Scroll_Depth', 'Session_Duration']

results = []
for k in list_criteria:
    light_mean_res = df.loc[df['Theme'] == 'Light Theme', k].mean()
    dark_mean_res = df.loc[df['Theme'] == 'Dark Theme', k].mean()

    results.append({'Metric': k,
                    'Light Theme Mean': light_mean_res,
                    'Dark Theme Mean': dark_mean_res,
                    'Dynamic': round((dark_mean_res - light_mean_res)/abs(light_mean_res)*100,3)})

    print(f'Light Theme (H0): {k} = {light_mean_res:.4f}')
    print(f'Dark Theme (H1): {k} = {dark_mean_res:.4f}')

df1 = pd.DataFrame(results)
```

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = data[((data < lower) | (data > upper))]
    return outliers

def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - data.mean()) / data.std()
    outliears = data[abs(z_scores) > threshold]
    return outliears

iqr_bmi = detect_outliers_iqr(ins_df['bmi'])
iqr_charges = detect_outliers_iqr(ins_df['charges'])

zscore_bmi = detect_outliers_zscore(ins_df['bmi'])
zscore_charges = detect_outliers_zscore(ins_df['charges'])
```

На даному етапі я стикнувся із проблемою виявлення нових викидів, так як Z-Score та IQR методи базуються на відстанях між точками і при заміні найдовших точок, після оновлення даних - методи шукають найдовші точки (у нових даних), які не відповідають загальній масі. Тому варто б визначити більш-менш оптимальні ваги для кожного методу, коли дані будуть "стабільні" і не буде виявлено нових викидів. Далі:

```python
def analyze_outliers_with_thresholds(data, column_name, iqr_multipliers=[1.5, 2, 2.5, 3], z_thresholds=[2, 2.5, 3, 3.5]):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # IQR аналіз
    iqr_results = []
    for mult in iqr_multipliers:
        Q1 = data[column_name].quantile(0.25)
        Q3 = data[column_name].quantile(0.75)
        IQR = Q3 - Q1
        lower, upper = Q1 - mult * IQR, Q3 + mult * IQR
        outliers = data[column_name][(data[column_name] < lower) | (data[column_name] > upper)]
        iqr_results.append({
            'multiplier': mult,
            'n_outliers': len(outliers),
            'outliers': outliers
        })
        ax1.scatter(outliers.index, outliers.values, label=f'IQR {mult}x ({len(outliers)} outliers)')

    # Z-score аналіз
    zscore_results = []
    for threshold in z_thresholds:
        z_scores = (data[column_name] - data[column_name].mean()) / data[column_name].std()
        outliers = data[column_name][abs(z_scores) > threshold]
        zscore_results.append({
            'threshold': threshold,
            'n_outliers': len(outliers),
            'outliers': outliers
        })
        ax2.scatter(outliers.index, outliers.values, label=f'Z-score {threshold} ({len(outliers)} outliers)')

    # Налаштування графіків
    ax1.plot(data.index, data[column_name], alpha=0.2, color='gray', label='All data')
    ax2.plot(data.index, data[column_name], alpha=0.2, color='gray', label='All data')

    ax1.set_title('IQR Method Outliers')
    ax2.set_title('Z-score Method Outliers')

    ax1.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)
    ax2.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)

    plt.tight_layout()
    plt.suptitle(f'Testing multiple parameters for outlier detection in "{column_name.capitalize()}"', size=16, y=1.05)
    plt.show()

    # Вивід результатів
    print("\nIQR Method Results:")
    for r in iqr_results:
        print(f"Multiplier: {r['multiplier']}, Outliers: {r['n_outliers']}")

    print("\nZ-score Method Results:")
    for r in zscore_results:
        print(f"Threshold: {r['threshold']}, Outliers: {r['n_outliers']}")

# Використання
analyze_outliers_with_thresholds(cleaned_ins, 'charges',
                                iqr_multipliers=[1.5, 2, 2.5, 3],
                                z_thresholds=[2, 2.5, 3, 3.5])
```

```python
def fill_by_chunks(df, target_col, reference_col, window=5):
    # подумав, що можливо краще заповнювати пропущені дані порівнючи чанки (вікна) у яких найбільша кореляція у референтній колонці
    # або ж просто можна було не запарюватись (частка ) пропущених значення суттєво мала - можна було просто дропнути
    result = df.copy()
    missing = df[target_col].isna()

    # Заповнюємо пропуски
    for idx in df[missing].index:
        ref_val = df.loc[idx, reference_col]
        similar = df[
            (df[reference_col].between(ref_val - window, ref_val + window)) &
            (df[target_col].notna())
        ][target_col]
        result.loc[idx, target_col] = similar.median() if len(similar) > 0 else ref_val

    return result

df_filled_train = fill_by_chunks(df_train, 'Arrival Delay in Minutes', 'Departure Delay in Minutes', window=5)
print(f'Correlation between arrival and departure delay after filling missing values: {round(df_filled_train['Arrival Delay in Minutes'].corr(df_filled_train['Departure Delay in Minutes']),4)}')
```

```python
from factor_analyzer import FactorAnalyzer
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo

scaler = StandardScaler()
X = df_full_filled.iloc[:,7:23]
X_scaled = scaler.fit_transform(X)

kmo_all, kmo_model = calculate_kmo(X_scaled)
chi_square_value, p_value = calculate_bartlett_sphericity(X_scaled)

print(f'KMO: {kmo_model:.3f}')
print(f'Test Bartelett p-value: {p_value:.3f}')

# Create FactorAnalyzer object
fa = FactorAnalyzer(rotation="varimax")

# Find optimal number of factors using eigenvalues
fa.fit(X_scaled)
ev, v = fa.get_eigenvalues()

# Plot eigenvalues (scree plot)
plt.figure(figsize=(8,5))
plt.plot(range(1, len(ev)+1), ev, marker='o')
plt.xlabel('Number of Factors')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.grid(True, alpha=0.5)
plt.show()

# Select number of factors where eigenvalues > 1 (or elbow point from scree plot)
n_factors = sum(ev > 1)
print(f"Number of factors to extract: {n_factors}")

# Run EFA with selected number of factors
fa = FactorAnalyzer(n_factors=n_factors, rotation="varimax")
fa.fit(X_scaled)

# Get loadings
loadings = pd.DataFrame(fa.loadings_, index=X.columns)
print(loadings)

# Optional: visualize factor loadings
plt.figure(figsize=(10,6))
sns.heatmap(loadings, cmap="coolwarm", annot=True)
plt.title('Factor Loadings Heatmap')
plt.show()
```
