# Task about lakes
- Generate 20 random variable from 1 to 100
- Draw the plot of the sequence
- Lets assume each point represent the height and so all plotting is 2d mountains.
- Then consider the unlimited rain from above - cavities become lakes full of water.
- Determine the deepest lake

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
a = np.random.randint(1, 100, 20)
```

```python
# Функції для пошуку найглибшого озера
def is_lake_segment(values, i, j):
    left, right = values[i], values[j]
    middle = values[i+1:j]
    if left > max(middle) and right > max(middle):
        depth = min(left, right) - min(middle)
        return True, depth
    return False, 0


def find_deepest_lake(values):
    max_depth = 0
    lake_indices = None
    n = len(values)
    for i in range(n - 2):
        for j in range(i + 2, n):
            valid, depth = is_lake_segment(values, i, j)
            if valid and depth > max_depth:
                max_depth = depth
                lake_indices = (i, j)
    return max_depth, lake_indices


# Знаходимо найглибше озеро
depth, (l, r) = find_deepest_lake(a)
lake = a[l:r+1]
```

```python
# Візуалізація
plt.figure(figsize=(10, 5))
plt.plot(a, label='Висоти')
plt.plot(range(l, r+1), lake, c='red', label='Найглибше озеро')
plt.title("Графік висот з виділеним найглибшим озером")
plt.xlabel("Індекс")
plt.ylabel("Висота")
plt.axhline(a.max(), color="g", linestyle="dashed")
plt.axhline(a.min(), color="g", linestyle="dashed")
plt.legend()
plt.show()

# Виводимо результати
print(f"Глибина найглибшого озера: {depth}")
print(f"Індекси озера: l = {l}, r = {r}")
```
